```python
# Explode the 'platforms' column
exploded_df = data_La.explode('platforms')

# Create a new DataFrame with each platform value on a separate row
new_data = {
    'platforms': [],
    'attack_id': []
}

for index, row in exploded_df.iterrows():
    platforms = row['platforms']
    attack_id = row['attack_id']
    for platform in platforms:
        new_data['platforms'].append(platform)
        new_data['attack_id'].append(attack_id)

new_df = pd.DataFrame(new_data)

# Display the transformed DataFrame
print(new_df)
```
